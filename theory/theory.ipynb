```python
import pandas as pd
import re
```

```python
df=pd.read_table('soper',header=None)
df.columns=['theory']
df
```

```python
df['theory']=df.theory.apply(lambda x: re.sub(r'\(.+?\)', '',x) )
```

```python
def addTheory(theory):
    text='''    
---start---
\'\'\'this_title\'\'\'
{{theory papers}}
[[Category:theory]]
---end---
'''
    text= text.replace('this_title',theory)
    theories_file.write(text)
```

```python
theories_file='theories.csv'
theories_file=open(theories_file,'a')
for i,row in df.iterrows():
    theory=row['theory']
    addTheory(theory)
```

```python
theories_file.close()
```
