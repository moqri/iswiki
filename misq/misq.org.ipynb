```python
import urllib2
from bs4 import BeautifulSoup

url="http://misq.org/catalogsearch/advanced/result/?misq_year=20&limit=all"
response = urllib2.urlopen(url)
html = response.read()
```

```python
soup = BeautifulSoup(html, 'html.parser')
links=soup.find_all('a')
links=[i.get('href') for i in links]
links=[i for i in links if ("http://misq.org/" in i and "-" in i)]
links=list(set(links))
len(links)
```

```python
for link in links:
    print link
    response = urllib2.urlopen(link)
    page = response.read()
    title=link.replace('http://misq.org/','').replace('catalog/product/view/id/1528/s/','')
    f = open(title, 'w')
    f.write(page) 
    f.close()
```
